```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
df = pd.read_csv(r"C:\Users\aanak\Downloads\iris\Iris.csv")
df.head()
```

```python
df. rename(columns = {'SepalLengthCm':'function1', 'SepalWidthCm':'function2', 'PetalLengthCm':'function3', 'PetalWidthCm':'function4'}, inplace = True)
```

```python
df.head()
```

```python
df=df.replace(to_replace ="Iris-setosa",value ="syntax1")
```

```python
df=df.replace(to_replace ="Iris-virginica",value ="syntax2")
df=df.replace(to_replace ="Iris-versicolor",value ="syntax2")
df
```

```python
feature_cols = ["function1","function2","function3","function4"]
x = df[feature_cols]
y = df["Species"].map({"syntax1":0, "syntax2":1, "syntax3":2})
xtrain, xtest, ytrain, ytest = train_test_split(x, y)
clf = MLPClassifier(max_iter=600).fit(xtrain, ytrain)
clf.score(xtest,ytest)
```
